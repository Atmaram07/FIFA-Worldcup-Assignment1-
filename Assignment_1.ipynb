```python
#To show average attendence per year
at1 = df.groupby("Year")["Attendance"].mean().reset_index()
at1["Year"] = at1["Year"].astype(int)
plt.figure(figsize=(12,7))
ax = sns.pointplot(at1["Year"],at1["Attendance"],color="b")
ax.set_facecolor("grey")
plt.title("Average attendence by year",color='b')
plt.show()
```

```python
#total matches played and qualified teams per year
plt.figure(figsize=(12,7))
sns.pointplot(wc["Year"],wc["MatchesPlayed"],linewidth=1,
            edgecolor="k"*len(wc),color="b",label="Total matches played")
sns.pointplot(wc["Year"],wc["QualifiedTeams"],linewidth=1,
            edgecolor="k"*len(wc),color="r",label="Total qualified teams")
plt.title("Qualified teams by year",color='b')
plt.grid(True)
plt.ylabel("total matches and qualified teams by year")
plt.show()
```

```python
#Cities that hosted highest world cup matches
mt_cu = df["City"].value_counts().reset_index()
plt.figure(figsize=(10,8))
ax = sns.pointplot(y=mt_cu["index"][:15],x = mt_cu["City"][:15],
                 linewidth=1)
plt.xlabel("number of matches")
plt.ylabel("City")
plt.grid(True)
plt.title("Cities with maximum world cup matches",color='b')

plt.show()
```

```python
#Average attendance by city
cty_at = df.groupby("City")["Attendance"].mean().reset_index()
cty_at = cty_at.sort_values(by="Attendance",ascending=True)

plt.figure(figsize=(15,5))

axis = sns.pointplot("Attendance","City",
            data=cty_at[:15],
            linewidth =1 )

plt.title("Average attendance by city",color='b')
plt.show()

```

```python
#World cup final results by nation

cou = wc["Winner"].value_counts().reset_index()
cu_win = cou.copy()
cu_win.columns = ["country","count"]
cu_win["type"] = "WINNER"


plt.figure(figsize=(10,10))
sns.pointplot("count","country",data=cu_win,
            linewidth=1)

plt.title("Final results by nation",color='b')
plt.show()
```

```python
#Teams with highest fifa world cup goals
to_gl_h = df.groupby("Home Team Name")["Home Team Goals"].sum().reset_index()
to_gl_h.columns = ["team","goals"]

to_gl_a = df.groupby("Away Team Name")["Away Team Goals"].sum().reset_index()
to_gl_a.columns = ["team","goals"]

total_goals = pd.concat([to_gl_h,to_gl_a],axis=0)
total_goals = total_goals.groupby("team")["goals"].sum().reset_index()
total_goals = total_goals.sort_values(by="goals",ascending =False)
total_goals["goals"] = total_goals["goals"].astype(int)

plt.figure(figsize=(10,12))
ax= sns.pointplot("goals","team",data=total_goals[:20],
                linewidth=1,edgecolor="k"*20)


plt.title("Teams with highest fifa world cup goals",color='b')
plt.show()
```

```python
#Referee in most of the matches
rf = df["Referee"].value_counts().reset_index()
rf = rf.sort_values(by="Referee",ascending=False)

plt.figure(figsize=(10,10))
sns.barplot("Referee","index",data=rf[:15])
plt.xlabel("count")
plt.ylabel("Referee name")
plt.grid(True)
plt.title("Referee's with most matches",color='black')
plt.show()
```
